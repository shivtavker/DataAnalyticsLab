### Basic moments of one-dimensional data

```python
from scipy import stats

def compute_mean(arr):
    sum_array = 0
    for number in arr:
        sum_array += number
    return sum_array/len(arr)

def compute_median(arr):
    arr.sort()
    if(len(arr)%2 == 0):
        return 0.5*(arr[int(len(arr)/2)] + arr[int((len(arr)/2)) -1])
    else:
        return arr[(len(arr) -1)/2]

def compute_mode(arr):
    return stats.mode(arr)[0][0];

def compute_variance(arr):
    mean = compute_mean(arr)
    squared_sum = 0
    
    for number in arr:
        squared_sum += (number - mean)**2
    
    return squared_sum/(len(arr)-1) #----------- told by TA to use n-1 instead of n

def compute_stddev(arr):
    return (compute_variance(arr))**0.5

array = [1, 5, 8, 1, 2, 3, 4, 9, 11, 2, 3, 2, 2, 2]

print("Mean: ", compute_mean(array))
print("Median: ", compute_median(array))
print("Mode: ", compute_mode(array))
print("Variance: ", compute_variance(array))
print("Std_dev: ", compute_stddev(array))

```

### Feature visualisation and selection

Load the twodimdata.csv file. Comment on the following :
1. Are the features independent, why?
2. Can a feature be removed?

### Solution

- We see that features are linearly independent because covariance between X1, X2 turns out to be 0.
- However from the Scater plot it is clear that there exists some non-linear realtionship between X1 and X2.
- To figure out if there really exists a relationship, I have done a polynomial regression with X1 and X2 as columns of the matrix X and Y is column vector of a constant value(1).
- We see that Accuracy for above regression turns out to be 99.9999997%
- This shows that there exists a implicit relationship between X1 and X2. Since, the relationship is implicit. We cannot remove a feature 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import train_test_split

area = np.pi*3
colors = (0, 0, 0)
df = pd.read_csv('twodimdata.csv')
plt.scatter(df['x1'], df['x2'], s=area, c=colors, alpha=0.5)
plt.ylim(-0.003, 0.003)

clf = KernelRidge(alpha=1.0, kernel="poly", degree=2)

print("Covariance: ", np.cov(df['x1'], df['x2']))

X = np.column_stack((df['x1'], df['x2']))
Y = np.ones(len(df['x1']))


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

clf.fit(X_train, y_train)

def accuracy(array):
    difference_sum = 0
    for i in range(len(array)):
        difference_sum += (array[i] - y_test[i])**2
    return 1 - (difference_sum/len(array))

print("Accuracy: ", accuracy(clf.predict(X_test)))


```
