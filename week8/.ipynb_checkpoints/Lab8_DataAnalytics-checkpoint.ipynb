### Question 1:  
### Max Margin Classifier in 1D
**Let's assume two classes in 1D dimension. Class 1 : 9,10,11,12,13,14; Class -1 : -9,-10,-11,-12,-13,-14<br> <br>a) Plot these points. <br><br>b) Are they linearly separable?<br><br>c) Assume a threshold t. Points to one side of the threshold belong to class -1 and points to the other side of the threshold belong to class 1. Find the range of values of t which classifies the two classes completely. <br><br>d) Let's say y_i is the label and can take values -1 or 1. Max margin classifiers aims to find a supporting hyperplane (a boundary between the two classes) such the distance (or margin) from points belonging to either class is maximised. The points closest to the boundary on either side are termed as support vectors. Maximising the perpendicular distance from the support vectors to line will achieve the goal of max margin classifier. Our objective/cost function for the 1D case is cost = max{min(y_i(t-x_i))} where t is the threhold, y_i is the label and x_i is the data point. Plot the cost as a function of the threshold for the range of values of t found in the (c) part. <br><br>e) For what t is the cost maximum? This is the threshold that optimally classifies the two datasets. Now plot the threshold, and the two classes on a graph. Assign appropriate colours to distinguish the two classes and the threshold.<br><br>f) What are the support vectors?<br><br>g) Comment on the threshold. Is it the mid point of the two support vectors in this linearly separable 1D case?**

```python
class1 = np.array([9, 10, 11, 12, 13, 14])
classneg1 = np.array([-9, -10, -11, -12, -13, -14])

plt.scatter(class1, np.zeros(len(class1)), color='green')
plt.scatter(classneg1, np.zeros(len(classneg1)), color='red')
plt.title("Visualise the Data")
plt.show()

print("Yes we see that they are linearly seperable!!")

print("Any t between [-9, 9] will work as threshold!!")

threshold = np.linspace(-9, 9, num=200)
cost = []

for i in range(len(threshold)):
    cost.append(min(abs(threshold[i] - max(classneg1)), abs(threshold[i] - min(class1))))

plt.plot(threshold, cost)
plt.title("Cost vs Threshold")
plt.show()

plt.scatter(class1, np.zeros(len(class1)), color='green')
plt.scatter(classneg1, np.zeros(len(classneg1)), color='red')
plt.plot(np.zeros(10), np.linspace(-9, 9, num=10))
plt.title("Best Classifier!!")
plt.show()

print("point(-9) and point(9) are the support vectors")
print("Yes, the best threshold is the midpoint of the support vectors")
```

### SVM in 2D
**Using sklearn, visualise the data points and plot the decision boundary for the following 2D datapoints, into 2 classes. Also mention the support vectors for each case. Use sklearn's SVC implementation with 'linear' kernel and default C. <br> <br>
(a)  Class 1 - (-1,-2), (1,1), (1.5, 1.5), (-2,-1), (0,0), (1,-0.5) <br> Class 2 - (4.1,5.2), (6,8), (4,4), (7,8), (4.5,6), (5.5,6)**


```python
class0 = [[-1,2],[1,1],[1.5,1.5],[-2,-1],[0,0],[1,-0.5]]
class1 = [[4.1,5.2],[6,8],[4,4],[7,8],[4.5,6],[5.5,6]]

plt.scatter(*zip(*class0), color="red")
plt.scatter(*zip(*class1), color="green")
plt.title('Decision Boundary')

Y = [0]*len(class0) + [1]*len(class1)

clf = SVC(kernel = 'linear', gamma='auto')

clf.fit(class0 + class1, Y)

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2, 7.5)
yy = a * xx - (clf.intercept_[0]) / w[1]

plt.plot(xx, yy, 'k-')

print("Support Vectors: ")
print(clf.support_vectors_)

plt.scatter(*zip(*clf.support_vectors_), marker='D', s=75, facecolors='none', edgecolors='blue')
plt.xlim(-4, 8)
plt.ylim(-4, 8)
# point = [2.75, 2.75]
# plt.plot(xx, -xx + 5.5)
plt.show()
```

### Change of support vectors
**(d) How did the addition of one point in (b) and (c) affect the separating plane in (a)?**

### Effects on separating plane:

- We see that in part (b), the added point was closer to the hyperplane compared to preexisting support vectors. Hence, the support vectors changed and in turn we get a completely different hyperplane.

- In part (c) adding the extra point makes the data linearly not seperable and hence, hard margin SVM just fails to classify the Data. By changing it to soft margin SVM with C=1, we get a decision boundary.

```python
class0 = [[-1.3, -1.9], [-1,2],[1,1],[1.5,1.5],[-2,-1],[0,0],[1,-0.5]]
class1 = [[4.1,5.2],[6,8],[4,4],[7,8],[4.5,6],[5.5,6]]

plt.scatter(*zip(*class0), color="red")
plt.scatter(*zip(*class1), color="green")
plt.title('Decision Boundary')

Y = [0]*len(class0) + [1]*len(class1)

clf = SVC(kernel = 'linear', gamma='auto')

clf.fit(class0 + class1, Y)

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2, 7.5)
yy = a * xx - (clf.intercept_[0]) / w[1]

plt.plot(xx, yy, 'k-')

print("Support Vectors: ")
print(clf.support_vectors_)

plt.scatter(*zip(*clf.support_vectors_), marker='D', s=75, facecolors='none', edgecolors='blue')
plt.xlim(-4, 8)
plt.ylim(-4, 8)
# point = [2.75, 2.75]
# plt.plot(xx, -xx + 5.5)
plt.show()

print("We see that the Decision Boundary has not changed at all!!")
print("This is because the new point is farther away from the seperating hyperplane")
print("and does not alter the support vectors")
```

### Perturbation of non-support vectors
**(f) Now perturb all the points in both classes in part (a) except (1.5,1.5) in class0 and (4,4) in class1 as follows. Add -0.5 to each x and y coordinate in Class0 and 0.5 to each point in Class1. Did the decision boundary change?**

```python
class0 = [[-1,2],[1,1],[1.5,1.5],[-2,-1],[0,0],[1,-0.5]]
class1 = [[4.1,5.2],[6,8],[4,4],[7,8],[4.5,6],[5.5,6]]

for i in range(len(class0)):
    if(class0[i] != [1.5, 1.5]):
        class0[i] = [class0[i][0] + 0.5, class0[i][1] + 0.5]

for i in range(len(class1)):
    if(class1[i] != [4, 4]):
        class1[i] = [class1[i][0] + 0.5, class1[i][1] + 0.5]

plt.scatter(*zip(*class0), color="red")
plt.scatter(*zip(*class1), color="green")
plt.title('Decision Boundary')

Y = [0]*len(class0) + [1]*len(class1)

clf = SVC(kernel = 'linear', gamma='auto')

clf.fit(class0 + class1, Y)

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2, 7.5)
yy = a * xx - (clf.intercept_[0]) / w[1]

plt.plot(xx, yy, 'k-')

print("Support Vectors: ")
print(clf.support_vectors_)

plt.scatter(*zip(*clf.support_vectors_), marker='D', s=75, facecolors='none', edgecolors='blue')
plt.xlim(-4.5, 8.5)
plt.ylim(-4.5, 9.5)
# point = [2.75, 2.75]
# plt.plot(xx, -xx + 5.5)
plt.show()

print("We see that the Decision Boundary has not changed!!")
```

**(g) How did the perturbation of points which are not support vectors in (a) affect the final classifier? Did the decision boundary change? Why not?**

### Observations and Reasons:

- We see that perturbation of points which are not the support vectors does not affect the final classifier.
- This is mainly because the given perturbation just moves the points of each class further away from the Decision Boundary and hence, the support vectors remain unaltered.
- Incase, we have some different perturbation which moves some points closer to the Decision Boundary, Our Support Vectors will change and so does our decision boundary. This happened in part(b).

### Question 4  
### Introduction of Kernels
**a) Plot two concentric circles of radii 4 and 8 with origin at 0. Sample 100 points uniformly from each of these circles. (use rcos(theta) and rsin(theta) to find the coordinates of the points lying on the circle. r is the radius of the circle. Vary theta from 3.6 degrees to 360 degrees in intervals of 3.6 degrees.)<br><br>b) Is the data linearly separable? <br><br>c)Since it's not, we will transform them to a different space (with probably different number of dimensions). The transformation (which generally boils down to inner product - something that we call a kernel) will make the data linearly separable in the different space. For each point, compute features as [1, x, x^2,xy,y^2,y]. This transforms the data from 2D to 6D. Now, use sklearn's SVM classifier (linear kernel) to classify the transformed feature space into two classes. <br><br>d)Now apply SVM classifier with i) polynomial kernel with degree 2  ii) rbf kernel and perform the classification. Visualize the decision boundary in the original 2D space in both the cases. Report the accuracies. Which performs best and why?<br>**

```python
class0 = [] # radius 4
class1 = [] # radius 8

for i in range(100):
    theta = (math.pi/180)*3.6*i
    class0.append([4*math.cos(theta), 4*math.sin(theta)])
    class1.append([8*math.cos(theta), 8*math.sin(theta)])

plt.scatter(*zip(*class0), color="red")
plt.scatter(*zip(*class1), color="green")
plt.title('Visualise Data')
plt.show()

print("We see that the data is not linearly seperable")
print()

Y = [0]*len(class0) + [1]*len(class1)

clf = SVC(kernel = 'linear', gamma='auto')

clf.fit(class0 + class1, Y)

print("Accuracy Score without transformation: ", clf.score(class0 + class1, Y))

new_class0 = []
new_class1 = []

for i in range(len(class0)):
    x_0 = class0[i][0]
    y_0 = class0[i][1]
    x_1 = class1[i][0]
    y_1 = class1[i][1]
    
    new_class0.append([1, x_0, x_0**2, x_0*y_0, y_0**2, y_0])
    new_class1.append([1, x_1, x_1**2, x_1*y_1, y_1**2, y_1])

clf = SVC(kernel = 'linear', gamma='auto')

clf.fit(new_class0 + new_class1, Y)

print("Accuracy Score with transformation: ", clf.score(new_class0 + new_class1, Y))

print()

print("Accuracy Score of 1 on the training set implies")
print("that we have made the data linearly seperable and have classified it correctly")


#------------ KErnels --------------

clf = SVC(kernel='poly', gamma='auto', degree=2)
clf.fit(class0+class1, Y)

fig, ax = plt.subplots()
# title for the plots
title = ('Decision surface of Poly SVM Kernel')
# Set-up grid for plotting.
X0, X1 = -8, 8
xx, yy = np.meshgrid(np.arange(-9, 9, 0.1), np.arange(-9, 9, 0.1))

# print(clf.predict([[2, 0]]))
# plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

Color = []
for i in range(len(Z)):
    for j in range(len(Z[i])):
        if(Z[i][j] == 0):
            Color.append('#ad4551')
        else:
            Color.append("#5db072")

ax.scatter(xx, yy, c= Color, cmap=plt.cm.coolwarm, s=10)
ax.scatter(*zip(*class0), color="red")
ax.scatter(*zip(*class1), color="green")
ax.set_ylabel('X axis')
ax.set_xlabel('Y axis')
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(title)
plt.show()

clf = SVC(kernel='rbf', gamma='scale')
clf.fit(class0+class1, Y)

fig, ax = plt.subplots()
# title for the plots
title = ('Decision surface of RBF SVM Kernel')
# Set-up grid for plotting.
X0, X1 = -8, 8
xx, yy = np.meshgrid(np.arange(-9, 9, 0.1), np.arange(-9, 9, 0.1))

# print(clf.predict([[2, 0]]))
# plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

Color = []
for i in range(len(Z)):
    for j in range(len(Z[i])):
        if(Z[i][j] == 0):
            Color.append('#ad4551')
        else:
            Color.append("#5db072")

ax.scatter(xx, yy, c= Color, cmap=plt.cm.coolwarm, s=10)
ax.scatter(*zip(*class0), color="red")
ax.scatter(*zip(*class1), color="green")
ax.set_ylabel('X axis')
ax.set_xlabel('Y axis')
ax.set_xticks(())
ax.set_yticks(())
ax.set_title(title)
plt.show()
print()

print("Accuracy for both of the kernels is 100%")
print("RBF kernel is in general better because it is somewhat robust in terms of application")
print("For using Poly kernel, we have to know the exact degree to fit. For e.g if we use degree 3, it misclassifies")
print("For higher dimensional data, predicting the degree for fit, is not possible.")
print("For RBF Kernel we have to be careful with floating point accuracies.")
```

### Significance of Hyperparameters:

- Kernel: It is a transformation trick, we can evaluate and find the required parameters without actually transforming each data point to new Space. RBF kernel refers to Radial Basis Function. The resultant space for RBF and sigmoid kernels are Infinite Dimensional. Resultant Space for Polynomial Kernel has same dimensions as its degree. Poly Kernels represented as (1 + x1Tx2)**degree.

- C: This refers to the penalty. How much penalty do we incurr by misclassifying a training data point. Higher values of C implies Hard-Margin SVMs. Incase of outliers, Higher values of C are very unstable.

- Gamma: This is a parameter for RBF kernel. Intuitively, the gamma parameter defines how far the influence of a single training example reaches, with low values meaning 'far' and high values meaning 'close'. Very high values of gamma is similar to overfitting.

- Degree: Resultant Space for Polynomial Kernel has same dimensions as its degree. Poly Kernels are represented as (1 + x1Tx2)**degree.

```python
clf = LogisticRegression(random_state=0, solver='newton-cg', multi_class='multinomial').fit(small_x, small_y)
print("F1 Score Logistic Regression: ", f1_score(small_y, clf.predict(small_x), average='micro'))
print("Logistic Confusion Matrix: ")
print(confusion_matrix(small_y, clf.predict(small_x)))
print()
clf = KNeighborsClassifier(n_neighbors=5).fit(small_x, small_y)
print("F1 Score KNN: ", f1_score(small_y, clf.predict(small_x), average='micro'))
print("KNN Confusion Matrix: ")
print(confusion_matrix(small_y, clf.predict(small_x)))
print()
clf = GaussianNB().fit(small_x, small_y)
print("F1 Score NB: ", f1_score(small_y, clf.predict(small_x), average='micro'))
print("NB Confusion Matrix: ")
print(confusion_matrix(small_y, clf.predict(small_x)))
print()
print("We see that order for F1 Score is: Logistic > KNN > SVM > NB")
```

- Training time for SVM is lot more than Logistic Regression
- SVM minimizes hinge loss while logistic regression minimizes logistic loss.
