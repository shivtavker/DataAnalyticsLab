## Week 6: K-means Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

**(1a) Load the data file 'kmc-data1.xlsx' and visualize the data.**

```python
df = pd.read_excel('./kmc.xlsx', sheet_name="Data1", header= None)
X = df[0]
Y = df[1]
plt.scatter(X, Y, alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**(1b) Implement K-means clustering. **

**(1c) Let $K = 3$ and the initial cluster means be $[0,0], [4,4]$ and $[-4,4]$. Plot the final clusters with K-means clustering.**

```python
cluster_centres = [[2, 2], [5, 5], [3, 3]]
colors = ("green", "blue", "red")
no_cluster = len(cluster_centres)

no_iterations = 10;

clusterwise_points = []

for i in range(no_cluster):
    clusterwise_points.append([])

for i in range(len(X)):
    point = np.array([X[i], Y[i]])
    min_distance = np.linalg.norm(point - cluster_centres[0])**2
    min_centre = 0;
    for j in range(no_cluster):
        distance = np.linalg.norm(point - cluster_centres[j])**2
        if(distance < min_distance):
            min_centre = j
            min_distance = distance
    
    clusterwise_points[min_centre].append(point);  

for a in range(no_iterations):
    for j in range(no_cluster):
        cluster_centres[j] = np.mean(clusterwise_points[j], axis = 0)
        
    for k in range(no_cluster):
        clusterwise_points[k] = []
    
    for m in range(len(X)):
        point = np.array([X[m], Y[m]])
        min_distance = np.linalg.norm(point - cluster_centres[0])**2
        min_centre = 0;
        for j in range(no_cluster):
            distance = np.linalg.norm(point - cluster_centres[j])**2
            if(distance < min_distance):
                min_centre = j
                min_distance = distance
        clusterwise_points[min_centre].append(point);
    if(a < 5):
        for i in range(no_cluster):
            points = clusterwise_points[i];
            X_scatter = []
            Y_scatter = []
            for point in points :
                X_scatter.append(point[0]);
                Y_scatter.append(point[1]);

            plt.scatter(X_scatter, Y_scatter, color = colors[i], alpha = 0.5)

        for i in range(no_cluster):
            plt.scatter(cluster_centres[i][0], cluster_centres[i][1], color = colors[i], alpha = 1, s = 500)
        plt.title('After ' + str(a+1) + ' iterations')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.show()

clusterwise_points = np.array(clusterwise_points);        

for i in range(no_cluster):
    points = clusterwise_points[i];
    X_scatter = []
    Y_scatter = []
    for point in points :
        X_scatter.append(point[0]);
        Y_scatter.append(point[1]);
        
    plt.scatter(X_scatter, Y_scatter, color = colors[i], alpha = 0.5)
    
for i in range(no_cluster):
    plt.scatter(cluster_centres[i][0], cluster_centres[i][1], color = colors[i], alpha = 1, s = 500)

plt.title('Final Clusters')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**(1d) Compute the sum of square distance of each point from the cluster mean.**

```python
sum_distance = 0;

for i in range(no_cluster):
    for point in clusterwise_points[i]:
        sum_distance += np.linalg.norm(point - cluster_centres[i])**2

print("Sum of Square Distance of each point: ", sum_distance)
```

```python
initial2 = [[2, 2], [5, 5]];
initial3 = [[2, 2], [5, 5], [3, 3]];
initial4 = [[2, 2], [5, 5], [3, 3], [4, 4]];
array_centre = [initial2, initial3, initial4]

for cluster_centres in array_centre:
    colors = ("green", "blue", "red", "black")
    no_cluster = len(cluster_centres)

    no_iterations = 10;

    clusterwise_points = []

    for i in range(no_cluster):
        clusterwise_points.append([])

    for i in range(len(X)):
        point = np.array([X[i], Y[i]])
        min_distance = np.linalg.norm(point - cluster_centres[0])**2
        min_centre = 0;
        for j in range(no_cluster):
            distance = np.linalg.norm(point - cluster_centres[j])**2
            if(distance < min_distance):
                min_centre = j
                min_distance = distance

        clusterwise_points[min_centre].append(point);  

    for a in range(no_iterations):
        for j in range(no_cluster):
            cluster_centres[j] = np.mean(clusterwise_points[j], axis = 0)

        for k in range(no_cluster):
            clusterwise_points[k] = []

        for m in range(len(X)):
            point = np.array([X[m], Y[m]])
            min_distance = np.linalg.norm(point - cluster_centres[0])**2
            min_centre = 0;
            for j in range(no_cluster):
                distance = np.linalg.norm(point - cluster_centres[j])**2
                if(distance < min_distance):
                    min_centre = j
                    min_distance = distance
            clusterwise_points[min_centre].append(point);

    clusterwise_points = np.array(clusterwise_points);        

    for i in range(no_cluster):
        points = clusterwise_points[i];
        X_scatter = []
        Y_scatter = []
        for point in points :
            X_scatter.append(point[0]);
            Y_scatter.append(point[1]);

        plt.scatter(X_scatter, Y_scatter, color = colors[i], alpha = 0.5)

    for i in range(no_cluster):
        plt.scatter(cluster_centres[i][0], cluster_centres[i][1], color = colors[i], alpha = 1, s = 500)

    plt.title('Final Clusters for K = ' + str(no_cluster))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

### Comments:

- Whereas for K = 4, We have indentified a cluster(Blue) which doesn't have many data points and exists because of our initialisation. Black and Blue clusters and not well seperated.

**(2a) Load the data from the file 'kmc-data2.xlsx' and visualize the data. **

```python
df = pd.read_excel('./kmc.xlsx', sheet_name="Data2")
X = df['X']
Y = df['Y']
data = np.vstack((X, Y)).T
plt.scatter(X, Y, alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**(2b) Cluster the data with K-means clustering. Consider $K = 2$. **

```python
km = KMeans(
    n_clusters=2, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)

y_km = km.fit_predict(data)

plt.scatter(
    data[y_km == 0, 0], data[y_km == 0, 1], c='green', label='cluster 1'
)

plt.scatter(
    data[y_km == 1, 0], data[y_km == 1, 1], c='blue', label='cluster 2'
)

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.xlim(-2.0, 2.0)
plt.ylim(-2.0, 2.0)
plt.show()
```

**(2c) Suppose you want to cluster the data into two rings. Now, transform the data such that K-means clustering identifies the two rings as clusters. **

```python
Z = np.zeros(len(X));

for i in range(len(Z)):
    Z[i] = X[i]**2 + Y[i]**2
    
data_transform = np.vstack((X, Y, 5*Z)).T

km = KMeans(
    n_clusters=2, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)

y_km = km.fit_predict(data_transform)

plt.scatter(
    data_transform[y_km == 0, 0], data_transform[y_km == 0, 1], c='green', label='cluster 1'
)

plt.scatter(
    data_transform[y_km == 1, 0], data_transform[y_km == 1, 1], c='blue', label='cluster 2'
)

# plot the centroids
# print(km.cluster_centers_)

# plt.scatter(
#     km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
#     s=250, marker='*',
#     c='red', edgecolor='black',
#     label='centroids'
# )
plt.legend(scatterpoints=1)
plt.xlim(-2.0, 2.0)
plt.ylim(-2.0, 2.0)
plt.title("X - Y Plane")
plt.show()

# ------------- Y - Z data ----------------

plt.scatter(
    data_transform[y_km == 0, 1], data_transform[y_km == 0, 2], c='green', label='cluster 1'
)

plt.scatter(
    data_transform[y_km == 1, 1], data_transform[y_km == 1, 2], c='blue', label='cluster 2'
)

# plot the centroids
# print(km.cluster_centers_)

plt.scatter(
    km.cluster_centers_[:, 1], km.cluster_centers_[:, 2],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
# plt.xlim(-2.0, 2.0)
# plt.ylim(-2.0, 2.0)
plt.title("Y - Z Plane")
plt.show()


# ------------- X - Z data ----------------

plt.scatter(
    data_transform[y_km == 0, 0], data_transform[y_km == 0, 2], c='green', label='cluster 1'
)

plt.scatter(
    data_transform[y_km == 1, 0], data_transform[y_km == 1, 2], c='blue', label='cluster 2'
)

# plot the centroids
# print(km.cluster_centers_)

plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 2],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
# plt.xlim(-2.0, 2.0)
# plt.ylim(-2.0, 2.0)
plt.title("X - Z Plane")
plt.show()
```
