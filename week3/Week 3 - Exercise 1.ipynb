## Generate random variables

a) Generate 1000 independent samples from Normal distribution with mean $\mu = 0$ and variance $\sigma^2 = 1$. Plot and visualize the data.

(Hint: Generate Gaussian samples from uniform random variables.)


```python
import random
import matplotlib.pyplot as plt
from math import *

def generate_normal(mu, sigma):
    u = random.random()
    v = random.random()

    z1 = sqrt(-2 * log(u)) * sin(2 * pi * v)
    z2 = sqrt(-2 * log(u)) * cos(2 * pi * v)

    x1 = mu + z1 * sigma
    x2 = mu + z2 * sigma

    return x2

normal = []

for i in range(10000):
    normal.append(generate_normal(0, 1))

plt.hist(normal, normed=True, bins=40)

plt.ylabel('PDF')
plt.show()
```

### Generate correlated random variables

b) Generate 1000 independent samples from Bivariate Normal distribution with means $\mu_{X} = \mu_{Y} = 0$, variances $\sigma_{X}^2 = \sigma_{Y}^2 = 1$ and correlation coefficient $\rho = - 0.9$. Plot and visualize the data in 2D and 3D.

(Hint: If (X,Y) is Bivariate Normal with $\mu_X = \mu_Y = 0$, $\sigma_X^2 = \sigma^2_Y = 1$ and correlation coefficient $\rho$, then the conditional distribution of $Y$ (conditioned on $X = x$) is Normal with mean $\rho x$ and variance $1 - \rho^2$, i.e., $Y|_{X = x} \sim {\mathcal N}(\rho x, 1 - \rho^2)$.)


```python
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from math import *
from scipy.stats import multivariate_normal

def generate_normal(mu, variance):
    u = random.random()
    v = random.random()

    z1 = sqrt(-2 * log(u)) * sin(2 * pi * v)
    z2 = sqrt(-2 * log(u)) * cos(2 * pi * v)

    x1 = mu + z1 * sqrt(variance)
    x2 = mu + z2 * sqrt(variance)

    return x2

def bivariate_normal(mux, muy, varx, vary, rho):
    X = generate_normal(mux, varx)
    Y = generate_normal(rho*X, 1-(rho**2))
    return (X, Y)

pair_X = []
pair_Y = []
bivariate_array = []

for i in range(1000):
    bivariate = bivariate_normal(0, 0, 1, 1,-0.9)
    pair_X.append(bivariate[0])
    pair_Y.append(bivariate[1])
    bivariate_array.append(bivariate)
    
np_pair_X = np.array(pair_X)
np_pair_Y = np.array(pair_Y)

pos = np.empty(np_pair_X.shape + (2,))
pos[:, 0] = np_pair_X; pos[:, 1] = np_pair_Y    

plt.scatter(pair_X, pair_Y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y/X")
plt.show()

hist, xedges, yedges = np.histogram2d(pair_X, pair_Y, bins=14, range=[[-2, 2], [-2, 2]])

fig = plt.figure(figsize=[13,8])
ax = fig.add_subplot(111, projection='3d')



# Construct arrays for the anchor positions of the 16 bars.
xpos, ypos = np.meshgrid(xedges[:-1] , yedges[:-1] , indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Construct arrays with the dimensions for the 16 bars.
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

plt.show()
```

### Maximum Likelihood Estimate using Grid Search

e) Let ${\mathsf P}_{\rho}(\mathbf{X} = \mathbf{x}, \mathbf{Y} = \mathbf{y})$ denote the likelihood function (for the observation $(\mathbf{x}, \mathbf{y})$, as a function of the parameter $\rho$). Compute the likelihood function for the given data and find the maximum likelihood estimate of $\rho$, $\hat{\rho}_{ML}$, using grid search.

(Hint: Plot ${\mathsf P}_{\rho}(\cdot,\cdot)$ as a function of $\rho$.)

```python
from math import *
from scipy.stats import multivariate_normal
import random
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data_corr.csv')

X = df['X']
Y = df['Y ']

log_likelihood = [0]*99
rho_array = []

i=1

for i in range(1, 100):
    rho = ((2/100) * i) + -1.0
    rv = multivariate_normal([0, 0], [[1, rho], [rho, 1]])
    for j in range(len(X)):
        log_likelihood[i-1] += log(rv.pdf([X[j], Y[j]]))
    rho_array.append(rho)
    
index = np.argmax(log_likelihood)

print("Best Rho", ((2/100) * (index+1) + -1.0))
print("Max Loglikely hood",max(log_likelihood))

plt.scatter(rho_array, log_likelihood)
plt.show()
    




```

### Maximum Likelihood Estimate using Gradient Descent

f) Find the maximum likelihood estimate of $\rho$, $\hat{\rho}_{ML}$, using gradient descent. Plot the estimate of $\rho$ (versus iteration) for different choices of step size (e.g., 0.001, 0.01, 0.1).

```python
import pandas as pd
import matplotlib.pyplot as plt

def compute_gradient(x, y, rho):
    gradient = -2*rho/(1-(rho**2)) + 2*rho*(x**2 + y**2)/((1-(rho**2))**2) -2*(1+(rho**2))*x*y/((1-(rho**2)))**2
    return gradient

rho = 0

step_size = [0.001, 0.01, 0.1]

df = pd.read_csv('data_corr.csv')

X = df['X']
Y = df['Y ']

for step in step_size:
    rho = 0
    rho_array = [rho]
    iteration = [0]
    for i in range(max(int(1/step), 100)):
        total_gradient = 0
        for j in range(len(X)):
            total_gradient += compute_gradient(X[j], Y[j], rho)
        if(total_gradient < 0):
            rho += step
        elif(total_gradient > 0):
            rho += -step
        elif(total_gradient == 0):
            break
        rho_array.append(rho)
        iteration.append(i)
    plt.figure(figsize=(15,6))
    plt.scatter(iteration, rho_array)
    plt.plot(iteration, rho_array)
    plt.xlabel("Iteration")
    plt.ylabel("rho_value")
    plt.title("Step Size: " + str(step))
    plt.show()
    print("Best Rho: " + str(rho) + "For step size: " + str(step))
```

### Maximum Likelihood Estimate (Analysis)

g) Find the maximum likelihood estimate of $\rho, \hat{\rho}_{ML}$ analytically (from the likelihood function).

```python
#------- For Multivariate Normal the maximum likelihood estimate of covariance matrix 
##comes out to be same as sample covariance matrix
import pandas as pd
import numpy as np

df = pd.read_csv('data_corr.csv')

X = df['X']
Y = df['Y ']

rho = np.cov(X, Y)[0][1] * (len(X)-1)/len(X)

print("Maximum Likelihood Estimate(biased) of Rho: ", rho)
```

### Polynomial Regression using Scikit-learn Libraries

Fit a polynomial model (for the suggested degree) to the data. Comment on the fit (as a function of the model) in terms of the mean square error.

(Hint: Use in-built libraries from Scikit-learn for the exercise.)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

data = np.load('data_poly1.npz')
lst = data.files
X = []
Y = []

for items in lst:
    if items == 'x':
        X = data[items]
    else:
        Y = data[items]
        
X = np.array(X);
X = X.reshape(-1, 1)
```

### Degree 1 (Linear)

j) Using linear regression, seek a relation between $Y$ and $X$. Plot the function and report the mean square error.

```python
reg = LinearRegression().fit(X, Y)
plt.scatter(X, Y, color = 'blue') 
plt.plot(X, reg.predict(X), color = 'red') 
plt.title('Linear Regression') 
plt.xlabel('Temperature') 
plt.ylabel('Pressure') 
plt.show()

print("Mean-Squared Error: ", mean_squared_error(Y, reg.predict(X)))
```

### Degree 2 (Quadratic)

k) Using polynomial regression of degree 2, seek a relation between $Y$ and $X$. Plot the regression function (vs $X$) and report the mean square error.

```python
poly = PolynomialFeatures(degree = 2) 
X_2_poly = poly.fit_transform(X)  
reg2 = LinearRegression() 
reg2.fit(X_2_poly, Y)

X_plot = np.arange(-10,11,1).reshape(-1,1)

plt.scatter(X, Y, color = 'blue')
plt.plot(X_plot, reg2.predict(poly.fit_transform(X_plot)), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Temperature') 
plt.ylabel('Pressure') 
  
plt.show()

print("Mean-Squared Error: ", mean_squared_error(Y, reg2.predict(X_2_poly)))
```

### Degree 9

l) Using polynomial regression of degree 9, seek a relation between $Y$ and $X$. Plot the regression function (vs $X$) and report the mean square error.

```python
poly = PolynomialFeatures(degree = 9) 
X_9_poly = poly.fit_transform(X)  
reg3 = LinearRegression() 
reg3.fit(X_9_poly, Y)

X_plot = np.arange(-10,11,1).reshape(-1,1)

plt.scatter(X, Y, color = 'blue')
plt.plot(X_plot, reg3.predict(poly.fit_transform(X_plot)), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Temperature') 
plt.ylabel('Pressure') 
  
plt.show()

print("Mean-Squared Error: ", mean_squared_error(Y, reg3.predict(X_9_poly)))
```

### Training Error

m) Plot the mean square error versus the degree of the polynomial fit (for degree = 0,1, ..., 9). Comment on the best model for the data.

```python
train_error = []
for i in range(0,10):
    poly = PolynomialFeatures(degree = i) 
    x_poly = poly.fit_transform(X)  
    reg = LinearRegression() 
    reg.fit(x_poly, Y) 
    train_error += [mean_squared_error(Y, reg.predict(poly.fit_transform(X)))]
plt.plot(train_error)
plt.xlabel("Degree")
plt.ylabel("Mean Squared Error")
plt.ylim(0, 100)
plt.xlim(0, 12)
plt.show()

print()
print("We see that training error decreases as we increase the degree of polynomial fit")
print("If we just see the train error we might conclude that degree 9 fit is the best.")
print("However, there is a high chance that we might overfit.")
print("Therefore we need a validation set to find the best model.")
```
