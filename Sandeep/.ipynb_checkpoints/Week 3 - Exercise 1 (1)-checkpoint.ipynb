## Generate random variables

a) Generate 1000 independent samples from Normal distribution with mean $\mu = 0$ and variance $\sigma^2 = 1$. Plot and visualize the data.

(Hint: Generate Gaussian samples from uniform random variables.)


```python
from numpy import random, sqrt,  log, sin, cos, pi
import matplotlib.pyplot as plt
# define a function to transform the randomly generated uniform random variables to a gaussian samples using Box- Muller transformation.
def gaussian_transformer(a,b):
    c = sqrt(-2*log(a))*cos(2*pi*b)
    d = sqrt(-2*log(a))*sin(2*pi*b)
    return(c,d)

# generating two independent samples from a uniform distribution
u1 = random.uniform(0,1,1000)
u2 = random.uniform(0,1,1000)

#transforming the generated samples to a gaussian sample using the transformation function to get two independent gaussian sample.
z1,z2 = gaussian_transformer(u1,u2)

#plotting the corresponding samples.
plt.hist(u1, bins = 100)
plt.show()
plt.hist(z1, bins = 100)
plt.show()
```

### Generate correlated random variables

b) Generate 1000 independent samples from Bivariate Normal distribution with means $\mu_{X} = \mu_{Y} = 0$, variances $\sigma_{X}^2 = \sigma_{Y}^2 = 1$ and correlation coefficient $\rho = - 0.9$. Plot and visualize the data in 2D and 3D.

(Hint: If (X,Y) is Bivariate Normal with $\mu_X = \mu_Y = 0$, $\sigma_X^2 = \sigma^2_Y = 1$ and correlation coefficient $\rho$, then the conditional distribution of $Y$ (conditioned on $X = x$) is Normal with mean $\rho x$ and variance $1 - \rho^2$, i.e., $Y|_{X = x} \sim {\mathcal N}(\rho x, 1 - \rho^2)$.)


```python
import numpy as np
import matplotlib.pyplot as plt
def conditional_distribution(a,rho):
    b = np.random.normal(rho*a, (1-rho**2))
    return(b)
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
y_given_x = conditional_distribution(x, -0.9)

joint = []
for i in range(1000):
    joint.append(y_given_x[i]*x[i])
a = np.array(x)
b = np.array(y_given_x)
c = np.array(joint)

plt.scatter(a,b, s=4)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y/X")
plt.show()

plt.hist2d(a,b,bins = 25)
plt.title("2D Histogram")
plt.xlabel("X")
plt.ylabel("Y/X")
plt.show()
```

## Estimate correlation coefficient

The file 'data_corr.csv' has 1000 independent samples from a Bivariate Normal distribution with means $\mu_X = \mu_Y = 0$, variances $\sigma_X^2 = \sigma_Y^2 = 1$ and unknown correlation coefficient $\rho$.

c) Load the data from the file 'data_corr.npz'. Plot and visualize the marginal and joint distributions of $X$ and $Y$.

d) Comment on the nature of correlation observed in the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
df =pd. read_csv("data_corr.csv")

X = df['X'].values
Y = df['Y '].values

plt.hist(X, bins = 25)
plt.title("Marginal Distribution X")
plt.show()

plt.hist(Y, bins = 25)
plt.title("Marginal Distribution Y")
plt.show()

plt.hist2d(X, Y, bins = 25)
plt.title("Joint Distribution")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar()
plt.show()

correlation = np.corrcoef(X,Y)
print(correlation)
print("Correlation Coefficient is: ", correlation[0][1])
# there is positive correlation between X and Y i.e. both increase or decrease together
```

### Simple Linear Regression

h) Using linear regression, seek a relation between $Y$ and $X$ (i.e., express $Y$ as a linear function of $X$). Comment.

```python
sig_X = np.sum(X)
sig_Y = np.sum(Y)
sig_X2 = 0
sig_XY = 0
n = len(X)
for i in X:
    sig_X2 = sig_X2 + i * i
for i in range (len(X)):
    sig_XY = sig_XY + X[i]*Y[i]
# let the equation be Y=aX +b
a = (n*sig_XY - sig_X*sig_Y)/(n*sig_X2 - sig_X**2)
b = (sig_Y*sig_X2 - sig_X*sig_XY)/(n*sig_X2 - sig_X**2)

print("The relation is: Y = ", a, '*X +', b )
```

## Polynomial Regression

Load the two dimensional data from the file 'data_poly1.npz'.

### Polynomial Regression using Scikit-learn Libraries

Fit a polynomial model (for the suggested degree) to the data. Comment on the fit (as a function of the model) in terms of the mean square error.

(Hint: Use in-built libraries from Scikit-learn for the exercise.)

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

```

```python
x1 = x1.reshape(-1,1)
y1 = y1.reshape(-1,1)
```

### Degree 1 (Linear)

j) Using linear regression, seek a relation between $Y$ and $X$. Plot the function and report the mean square error.

```python
lin = LinearRegression()
lin.fit(x1, y1)

plt.scatter(x1, y1, color = 'blue') 
plt.plot(x1, lin.predict(x1), color = 'red') 
plt.title('Linear Regression') 
plt.xlabel('Temperature') 
plt.ylabel('Pressure') 
  
plt.show()
```

### Degree 2 (Quadratic)

k) Using polynomial regression of degree 2, seek a relation between $Y$ and $X$. Plot the regression function (vs $X$) and report the mean square error.

```python
poly = PolynomialFeatures(degree = 2) 
x1_poly = poly.fit_transform(x1)  
lin2 = LinearRegression() 
lin2.fit(x1_poly, y1) 
```

```python
x1_new = np.arange(-10,11,1).reshape(-1,1)
```

```python
plt.scatter(x1, y1, color = 'blue')
plt.plot(x1_new, lin2.predict(poly.fit_transform(x1_new)), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Temperature') 
plt.ylabel('Pressure') 
  
plt.show() 
```

### Degree 9

l) Using polynomial regression of degree 9, seek a relation between $Y$ and $X$. Plot the regression function (vs $X$) and report the mean square error.

```python
poly = PolynomialFeatures(degree = 9) 
x1_poly = poly.fit_transform(x1)  
lin2 = LinearRegression() 
lin2.fit(x1_poly, y1) 

plt.scatter(x1, y1, color = 'blue')
plt.plot(x1_new, lin2.predict(poly.fit_transform(x1_new)), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Temperature') 
plt.ylabel('Pressure') 
  
plt.show() 
```

### Training Error

m) Plot the mean square error versus the degree of the polynomial fit (for degree = 0,1, ..., 9). Comment on the best model for the data.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
error_train = []
for d in range(0,10):
    poly = PolynomialFeatures(degree = d) 
    x_poly = poly.fit_transform(x1)  
    lin = LinearRegression() 
    lin.fit(x_poly, y1) 
    error_train += [mean_squared_error(y1,lin.predict(poly.fit_transform(x1)))]
plt.plot(error_train)
plt.xlabel("Degree")
plt.ylabel("Mean Squared Error")
plt.show()
    
```

### Test error

Load 'data_poly2.npz'. The data points in the file are from the same sample as 'data_poly1.npz'.

n) Plot the mean square error (for degrees 0,1,...,9) for the test data set, for the polynomial fit obtained with the training data set 'data_poly1.npz'. Now, comment on the best model for data.

(Hint: Plot Training error and Testing error in the same plot.)

```python
from numpy import load
import numpy as np
import matplotlib.pyplot as plt
dat1 = load('data_poly2.npz')
list1 = dat1.files
X1 = []
Y1 = []

for items in list1:
    if items == 'x':
        X1 = dat1[items]
    else:
        Y1 = dat1[items]
X1 = X1.reshape(-1,1)
Y1 = Y1.reshape(-1,1)

error_test = []
for d in range(0,10):
    poly = PolynomialFeatures(degree = d) 
    X_poly = poly.fit_transform(X1)  
    lin = LinearRegression() 
    lin.fit(X_poly, Y1) 
    error_test += [mean_squared_error(y1,lin.predict(poly.fit_transform(X1)))]
plt.plot(error_test)
plt.plot(error_train)
plt.xlabel("Degree")
plt.ylabel("Mean Squared Error")
plt.show()
```
